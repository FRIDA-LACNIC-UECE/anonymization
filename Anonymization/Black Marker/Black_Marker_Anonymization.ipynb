```python
import pandas as pd
import csv
```

```python
def open_df_pattern():
    """ Open the studied Dataframe and create a copy in standard path
    """
    
    df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",
                    names = ("Age", "Workclass", "fnlwgt", "Education", "Education_Number", "Marital_Status", "Occupation", "Relationship", "Race", "Sex", "Capital_Gain", "Capital_Loss", "Hours_Per_Week", "Native_Contry", "Annual_Gain"))
    df.to_csv('df_copy.csv', index=False)

    print("Raw Dataframe")
    display(df)
```

```python
def black_marker(column: str):
    """ Black Marking a specified column in the output csv dataframe
        :param - column: defines the column that will be anonymized
    """

    df = pd.read_csv('df_copy.csv')
    df.loc[:, column] = 'NULL'
    df.to_csv('df_copy.csv', index=False)
    
    print("Anonymized Dataframe")
    df_test = pd.read_csv('df_copy.csv')
    display(df_test)
```

```python
open_df_pattern()

black_marker("Workclass")
black_marker("Marital_Status")
```
